```python
d4_2015 = pd.read_csv("2015_atles_renda_index_gini.csv")
d4_2016 = pd.read_csv("2016_atles_renda_index_gini.csv")
d4_2017 = pd.read_csv("2017_atles_renda_index_gini.csv")
d4_2018 = pd.read_csv("2018_atles_renda_index_gini.csv")
d4_2019 = pd.read_csv("2019_atles_renda_index_gini.csv")
```

```python
# Step 1: Convert the column to numeric, coercing non-numeric values to NaN
d4_2015['Index_Gini'] = pd.to_numeric(d4_2015['Index_Gini'], errors='coerce')

# Step 2: Calculate the mean of the column, ignoring NaN values
mean_value = d4_2015['Index_Gini'].mean()

# Step 3: Replace NaN values (which were non-numeric) with the mean
d4_2015['Index_Gini'].fillna(mean_value, inplace=True)
```

```python
d4_2015_gr = d4_2015.groupby('Codi_Barri').mean("Index_Gini")
d4_2016_gr = d4_2016.groupby('Codi_Barri').mean("Index_Gini")
d4_2017_gr = d4_2017.groupby('Codi_Barri').mean("Index_Gini")
d4_2018_gr = d4_2018.groupby('Codi_Barri').mean("Index_Gini")
d4_2019_gr = d4_2019.groupby('Codi_Barri').mean("Index_Gini")
```

```python
if not pd.api.types.is_numeric_dtype(d4_2015['Index_Gini']):
    print("Index_Gini is not entirely numeric")
```

```python
df4 = pd.concat([d4_2015_gr, d4_2016_gr, d4_2017_gr, d4_2018_gr, d4_2019_gr])
df4_reset = df4.reset_index()
```

```python
df4_reset
```

```python
df4_reset.to_csv('/Users/svintitskyy/Desktop/uni/northwestern/classes/fall_23/cs333/a3/gini_index_15_19.csv', index=True)
```
