```python
import pandas as pd
import numpy as np
```

```python
birth_rates = pd.read_csv("birth_rates_15_19.csv").drop(columns=["Unnamed: 0", "index"])
birth_rates
```

```python
deaths = pd.read_csv("deaths_15_19.csv").drop(columns=["Unnamed: 0"])
deaths
```

```python
disp_income_household = pd.read_csv("disposable_income_household_15_19.csv").drop(columns=["Unnamed: 0"])
disp_income_household
```

```python
gini_index = pd.read_csv("gini_index_15_19.csv").drop(columns=["Unnamed: 0"])
gini_index
```

```python
gross_income = pd.read_csv("gross_income_15_19.csv")#.drop(columns=["Unnamed: 0"])
gross_income
```

```python
immigration = pd.read_csv("immigration_15_19.csv").drop(columns=["Unnamed: 0"])
immigration
```

```python
population = pd.read_csv("population_15_19.csv").drop(columns=["Unnamed: 0"])
population["Any"] = immigration["Any"]
```

```python
unemployment = pd.read_csv("unemployment_15_19.csv").drop(columns=["Unnamed: 0"])
unemployment
```

```python
tmp1 = pd.merge(birth_rates, deaths, on=['Codi_Barri', 'Any', 'Codi_Districte']).rename(columns={"Taxa_mil_hab":"Birth_Rates_Per1000"}).drop(columns=['AEB'])
tmp2 = pd.merge(tmp1, disp_income_household, on=['Codi_Barri', 'Any', 'Codi_Districte', 'Nom_Barri', 'Nom_Districte']).drop(columns=['Seccio_Censal']).rename(columns={"Valor":"Num_Deaths",
'Import_€_Any':'Avg_Household_Income'})
tmp3 = pd.merge(tmp2, gini_index, on=['Codi_Barri', 'Any', 'Codi_Districte']).drop(columns=['Seccio_Censal', 'NACIONALITAT_G'])
tmp4 = pd.merge(tmp3, gross_income, on=['Codi_Barri', 'Any', 'Codi_Districte']).rename(columns={"Import_Renda_Bruta_€":"Avg_Gross_Income"})
tmp5 = pd.merge(tmp4, immigration, on=['Codi_Barri', 'Any', 'Codi_Districte', 'Nom_Barri', 'Nom_Districte']).drop(columns=['index']).rename(columns={"Taxa_mil_hab":"Immigration_Per1000"})
tmp6 = pd.merge(tmp5, population, on=['Codi_Barri', 'Any', 'Codi_Districte']).drop(columns=['SEXE', 'AEB', 'Seccio_Censal']).rename(columns={"Valor":"Population"})
final_df = pd.merge(tmp6, unemployment, on=['Codi_Barri', 'Any', 'Codi_Districte']).drop(columns=['Mes']).rename(columns={"Pes_atur":"Unemployment_Rate"})
```
